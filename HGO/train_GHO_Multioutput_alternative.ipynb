```python
def plot_the_loss_curve(epochs, mse_training, mse_validation):

  plt.figure()
  plt.xlabel("Epoch")
  plt.ylabel("Loss")

  plt.plot(epochs[0:], mse_training[0:], label="Training Loss")
  plt.plot(epochs[0:], mse_validation[0:], label="Validation Loss")
  plt.legend()
  
  # We're not going to plot the first epoch, since the loss on the first epoch
  # is often substantially greater than the loss for other epochs.
  merged_mse_lists = mse_training[1:] + mse_validation[1:]
  highest_loss = max(merged_mse_lists)
  lowest_loss = min(merged_mse_lists)
  delta = highest_loss - lowest_loss
  print(delta)

  top_of_y_axis = highest_loss + (delta * 0.05)
  bottom_of_y_axis = lowest_loss - (delta * 0.05)
   
  plt.ylim([bottom_of_y_axis, top_of_y_axis])
  plt.show() 
```

```python
X = np.load('X_run_equi_12.npy')
y_uni = np.load('y_uni_run_12.npy')
y_bi = np.load('y_bi_run_12.npy')
```

```python
# Normalization of the input data
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)
```

```python
x_vec = np.array(y_uni[0,:,0])
```

```python
for i in range(len(y_uni)):
    y_diff = np.zeros(np.shape(y_uni[0]))
    for j in range(len(y_uni[0])-1):
        y_diff[j+1] = y_uni[i][j+1]-y_uni[i][j]
    #y_diff[0] = [1,0]
    y_uni[i] = y_diff

y_uni = y_uni[:,1:,1]
```

```python
for i in range(len(y_bi)):
    y_diff = np.zeros(np.shape(y_bi[0]))
    for j in range(len(y_bi[0])-1):
        y_diff[j+1] = y_bi[i][j+1]-y_bi[i][j]
    #y_diff[0] = [1,0]
    y_bi[i] = y_diff

y_bi = y_bi[:,1:,1:]
```

```python
y_bi_xx = y_bi[:,:,0]
y_bi_yy = y_bi[:,:,1]
```

```python
#let's save 10% of the data for testing. these curves won't be part of the model training 
test_split=0.1

#x_train, x_test,y_train,y_test = train_test_split(X, y, test_size = test_split)

x_train, x_test,y_uni_train,y_uni_test = train_test_split(X, y_uni, test_size = test_split)
x_train, x_test,y_bi_xx_train,y_bi_xx_test = train_test_split(X, y_bi_xx, test_size = test_split)
x_train, x_test,y_bi_yy_train,y_bi_yy_test = train_test_split(X, y_bi_yy, test_size = test_split)


#let's check the array shapes: 
print('Input data')
print('Training set: ',x_train.shape)
print('Test set: ',x_test.shape)
print(' ')
print('Features')
#print('Training set: ',y_train.shape)
#print('Test set: ',y_test.shape)
print('Uniaxial training set: ',y_uni_train.shape)
print('Uniaxial test set: ',y_uni_test.shape)
print(' ')
print('Equibiaxial xx training set: ',y_bi_xx_train.shape)
print('Equibiaxial xx test set: ',y_bi_xx_test.shape)
print(' ')
print('Equibiaxial yy training set: ',y_bi_yy_train.shape)
print('Equibiaxial yy test set: ',y_bi_yy_test.shape)
#train_size = train_df.shape[0]
#test_size = test_df.shape[0]

#reshape for keras training
#x_train = x_train.reshape((x_train.shape[0], x_train.shape[1], 2))
#x_test = x_test.reshape((x_test.shape[0], x_test.shape[1], 2))
#shuffle data
idx = np.random.permutation(len(x_train))
#x_train = x_train[idx]
#y_train = y_uni_train[idx]

y_uni_train = y_uni_train[idx]
y_bi_xx_train = y_bi_xx_train[idx]
y_bi_yy_train = y_bi_yy_train[idx]
```

```python
# Set the hyperparameters
VALIDATION_SPLIT = 0.2
LEARNING_RATE = 0.0001
BATCH_SIZE = 1024
EPOCHS = 20
INPUT_SHAPE = x_train.shape[1:]

inputs = keras.Input(shape=INPUT_SHAPE)
```

```python
def make_default_hidden_layers(inputs):
    dense = layers.Dense(128, activation="relu")
    x = dense(inputs)
    
  #  dropout = tf.keras.layers.Dropout(0.2)
  #  x = dropout(x)
    
    x = layers.Dense(128, activation="relu")(x)
    
    x = layers.Dense(128, activation="relu")(x)

    dropout = tf.keras.layers.Dropout(0.2)
    x = dropout(x)
    x = layers.Dense(9)(x)
    
    return(x)

def build_uniaxial_branch(inputs):
    
    x = make_default_hidden_layers(inputs)

    x = layers.Activation("linear", name="uniaxial")(x)
    return x
    
def build_equibiaxial_xx_branch(inputs):
    
    x = make_default_hidden_layers(inputs)
    
    x = layers.Activation("linear", name="equibiaxial_xx")(x)
    return x

def build_equibiaxial_yy_branch(inputs):
    
    x = make_default_hidden_layers(inputs)

    x = layers.Activation("linear", name="equibiaxial_yy")(x)
    return x


def assemble_full_model(input_shape):
    
    inputs = keras.Input(shape=input_shape)
    
    uniaxial_branch = build_uniaxial_branch(inputs)
    equibiaxial_xx_branch = build_equibiaxial_xx_branch(inputs)
    equibiaxial_yy_branch = build_equibiaxial_yy_branch(inputs)
    
    model = keras.Model(inputs=inputs,
                     outputs = [uniaxial_branch, equibiaxial_xx_branch, equibiaxial_yy_branch],
                     name="model")
    
    return model
```

```python
model = assemble_full_model(INPUT_SHAPE)

model.compile(
    loss = tf.keras.losses.Huber(),
    optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE),
#    metrics=["mean_squared_error"],
)

model.summary()
```

```python
tf.keras.utils.plot_model(model)
```

```python
callbacks = [keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)]

history=model.fit(
    x_train,
    [y_uni_train, y_bi_xx_train, y_bi_yy_train],
    validation_split=VALIDATION_SPLIT,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
)

test_loss_acc = model.evaluate(x_test, [y_uni_test,y_bi_xx_test,y_bi_yy_test])

print("Test accuracy", test_loss_acc)
```

```python
def plot_real_curves(k, test_curves):
    
    if test_curves.all() == y_uni_test.all():
        a = 0
    elif test_curves.all() == y_bi_xx_test.all():
        a = 1
    elif test_curves.all() == y_bi_yy_test.all():
        a = 2   
    else:
        raise ValueError("Test curves are not acceptable")
        
    
    pred_diff = model.predict(x_test[k:k+1])[a][0]
    pred = np.zeros([len(pred_diff)+1])
    pred[0] = 0
    for i in range(len(pred_diff)):
        pred[i+1] = pred_diff[i] + pred[i]

    plt.plot(x_vec,pred, "--",  color = "r", label = "prediction")

    
    true_dif = test_curves[k:k+1][0]
    true = np.zeros([len(true_dif)+1])
    true[0] = 0
    for i in range(len(pred_diff)):
        true[i+1] = true_dif[i] + true[i]

    plt.plot(x_vec,true, color = "g", label = "real")

    plt.xlabel("λ")
    plt.ylabel("σ")
    plt.legend()
    plt.show()
```

```python
def rel_error(true,guess):
    err = np.abs(true-guess)/np.abs(true)
    return err
```

```python
def convert_back_curves(y_diff):
    ### Transforms the model's output (which is written in terms of differences between points) back into absolute coordinates.
    
    abs_curves = np.zeros([np.shape(y_diff)[0],np.shape(y_diff)[1]+1])    
    for i in range(len(y_diff)):
        y_abs = np.zeros(np.shape(y_diff)[1]+1)
        for j in range(len(y_diff[0])):
            y_abs[j+1] = y_diff[i][j]+y_abs[j]
        abs_curves[i] = y_abs
    
    return abs_curves
```

```python
pred_uni_curves = convert_back_curves(model.predict(x_test)[0])
true_uni_curves = convert_back_curves(y_uni_test)

pred_bi_xx_curves = convert_back_curves(model.predict(x_test)[1])
true_bi_xx_curves = convert_back_curves(y_uni_test)

pred_bi_yy_curves = convert_back_curves(model.predict(x_test)[2])
true_bi_yy_curves = convert_back_curves(y_uni_test)
```

```python
def print_metrics(true_curves, pred_curves):

    # RELATIVE ERROR CRITERIA

    c = np.average(rel_error(true_curves[:,1:], pred_curves[:,1:]),axis = 1)

    print("Average relative error", np.round(np.average(c)*100,decimals = 2),"%")

    treshold = 0.05
    print("Relative error accuracy =",np.round(100*len(c[c<treshold])/len(c), decimals = 2),"%, for a treshold of",treshold*100,"%")
    print('\n')

    # R^2 SCORE CRITERIA
    d = r2_score(np.transpose(true_curves[:,1:]), np.transpose(pred_curves[:,1:]), multioutput='raw_values')

    print("Average R^2 score: ", np.round(np.average(d),decimals = 4))

    treshold = 0.95
    print("R^2 Accuracy =", np.round(100*len(d[d>treshold])/len(d),decimals=2), "%, for a treshold of", treshold)
    print('\n')

    # MSE CRITERIA
    mse = tf.keras.losses.MeanSquaredError(reduction=tf.keras.losses.Reduction.NONE)
    e = mse(true_curves[:,1:], pred_curves[:,1:]).numpy()

    print("Average MSE score: ", np.round(np.average(e),decimals = 2),'\n')
```

```python
print(' ','\033[1m' + "Unixaial metrics" + '\033[0m', '\n')
print_metrics(true_uni_curves, pred_uni_curves)

print(' ','\033[1m' + "Equibiaxial xx metrics" + '\033[0m', '\n')
print_metrics(true_bi_xx_curves, pred_bi_xx_curves)

print(' ','\033[1m' + "Equibiaxial yy metrics" + '\033[0m', '\n')
print_metrics(true_bi_yy_curves, pred_bi_yy_curves)
```
