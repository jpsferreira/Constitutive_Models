```python
import pandas as pd
import random
import sympy as sym
import numpy as np
from matplotlib import pyplot as plt

pd.set_option('display.max_rows', None)
```

```python
def ogden_uniaxial(params, stretch):
    #returns cauchy stress at yy direction
    
    #3x3 Identity Matrix
    I = sym.Matrix([[1,0,0],[0,1,0],[0,0,1]])

    #Symbolic Principal stretches.
    λ1 = sym.Symbol('λ1')
    λ2 = sym.Symbol('λ2')
    λ3 = sym.Symbol('λ3')
    
    #Symbolic Ogden Parameters
    
    #'α_k'
    α1 = sym.Symbol('α1')
    α2 = sym.Symbol('α2')
    α3 = sym.Symbol('α3')
    α = sym.Matrix([α1,α2,α3])
    
    #"μ_k"
    μ1 = sym.Symbol('μ1')
    μ2 = sym.Symbol('μ2')
    μ3 = sym.Symbol('μ3')
    μ = sym.Matrix([μ1,μ2,μ3])

    #Generate SEF (Strain Energy Function)
    #The Ogden model is based on this principal stretches, wheras the Mooney-Rivlin
    # was mostly based on the Invariants of C.
    # We will consider that F is in its principal direction, therefore, all matrices will be diagonal matrices.
    
    sef = 0
    for i in range(3):
        sef = sef + 2*(μ[i]/α[i]**2) * (λ1**α[i] + λ2**α[i] + λ3**α[i] - 3)
    
    #Second Piola Kirchoff Stresses 

    # Let Si be the principal Second Piola Kirchoff Stresses 
    # and Ci be the principal values of the Right Cauchy-Green Deformation Tensor 'C',
    # i = 1,2,3

    # Si = 2 * d(sef)/d(Ci) <=>
    # Si = 2 * d(sef)/d(λi) * d(λi)/d(Ci)  and  Ci = λi**2, so d(Ci)/d(λi) = 2*λi <=> dl/d(Ci) = 1/(2*λi) <=>
    # Si =  1/λi * d(sef)/dλ

    S1=(1/λ1)*sym.diff(sef,λ1)
    S2=(1/λ2)*sym.diff(sef,λ2)
    S3=(1/λ3)*sym.diff(sef,λ3)
    S = sym.Matrix([[S1,0,0], [0,S2,0], [0,0,S3]]) 
    
    #Deformation Gradient
    F=sym.Matrix([[λ1,0,0], [0,λ2,0], [0,0,λ3]])
    
    Ft=sym.transpose(F)  # Ft == F
  
    T=(F*S*Ft)  # Cauchy stresses with no BCs. It is easy to see that in this case Ti = Si * λi**2.
    
    # Boundary Conditions for a uniaxial load.
    T = T - (I*T[2,2])

    # For uniaxial loading of an incompressible material, the value of the principal extensions (λi) are:
    λ1v=1/sym.sqrt(stretch)
    λ2v=stretch
    λ3v=1/sym.sqrt(stretch)
    
    T = T.subs([(α1, params[0]), (α2, params[1]), (α3, params[2]), (μ1, params[3]), (μ2, params[4]), 
                                  (μ3,params[5]), (λ1, λ1v), (λ2, λ2v), (λ3, λ3v)])
    
    return stretch,T[1,1]

def get_curve(params,stretch_min,stretch_max,ninc):
    #stores Ogden uniaxial runs between minimum and a maximum stretch
    lst=[ogden_uniaxial(params, stretch) for stretch in np.linspace(stretch_min,stretch_max,ninc)]
    return lst
```

```python
# Let us aplly our function to a simbolic stretch and simbolic params.

α1 = sym.Symbol('α1')
α2 = sym.Symbol('α2')
α3 = sym.Symbol('α3')
α = sym.Matrix([α1,α2,α3])

μ1 = sym.Symbol('μ1')
μ2 = sym.Symbol('μ2')
μ3 = sym.Symbol('μ3')
μ = sym.Matrix([μ1,μ2,μ3])

params = sym.Matrix([α,μ]) # Symbolic params

λ = sym.Symbol('λ') # Symbolic Stretch

[stretch, stress] = ogden_uniaxial(params, λ)
sym.simplify(stress)

# Cauchy stress as a function of principal stretch (λ) and ogden parameters (α and μ).

# This is in conformity with results found in literature for incompressible, uniaxial loading.
# For example: https://doi.org/10.1002/msd2.12013 Section 2.2.2.
```

```python
#Initial data-----------------------------------------------------------------------------------------------------

amin=0.1     #min α_k value
amax=10.0    #max α_k value
umin=0.1     #min μ_k value
umax=2.0     #max μ_k value
decimals=2   #number of decimal cases of α_k and μ_k variables

n_a=5               #number of α vectors
n_u=5               #number of μ vectors
n = n_a * n_u      #number of combinations of vectors
#
st_max=3.0             #applied stretch
st_min=1.0
ninc=20        #number of stretch increments
#------------------------------------------------------------------------------------------------------------------
```

```python
# Generates random values for α and μ vectors
a_list = np.round(np.random.uniform(amin, amax, size=(n_a, 3)), decimals)
u_list = np.round(np.random.uniform(umin, umax, size=(n_u, 3)), decimals)
```

```python
# Computes the n combinations of α and μ vectors

params = np.array([])

for i in a_list:
    for j in u_list:
        for k in range(3):
            params = np.append(params, np.array([i[k],j[k]]))
params = params.reshape(-1, 6)
params
```

```python
# Append the combinations to a DataFrame

df = pd.DataFrame(params[:,0], columns = ['a1'])
df['a2'] = params[:,1]
df['a3'] = params[:,2]
df['u1'] = params[:,3]
df['u2'] = params[:,4]
df['u3'] = params[:,5]

df

# Later I will deal with duplicates
```

```python
#generate (x,y) data for each unique c10 at the dataframe
df['data']=df.apply(lambda x : get_curve(x,st_min,st_max,ninc), axis=1)
df
```
